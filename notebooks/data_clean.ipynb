```python
import pandas as pd
```

```python
mvps = pd.read_csv("../data/mvps.csv")
```

```python
mvps
```

```python
# Removing unneeded columns from MVP data

mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
mvps.head()
```

```python
# Read in players data

players = pd.read_csv("../data/players.csv")
```

```python
players
```

```python
# Removing unneeded columns from players data

del players["Unnamed: 0"]
del players["Rk"]
```

```python
players.head()
```

```python
# Removing asterisk from player names

players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.head(50)
```

```python
# Getting total stats for players that played for more than one team in a season

players.groupby(["Player", "Year"])
```

```python
def single_row(df):
    # Return if only one row (played for one team)
    if df.shape[0] == 1:
        return df
    # If multiple teams, use the total row with last team played for that year
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

```python
# Removing extra index levels

players.index = players.index.droplevel()
players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
# Merging mvp and player data

combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined.head()
```

```python
combined[combined["Pts Won"] > 0]
```

```python
# Replacing null values for players that did not recieve mvp votes

combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
# Reading in teams data

teams = pd.read_csv("../data/teams.csv")
```

```python
teams
```

```python
teams.head(50)
```

```python
# Removing rows with division

teams = teams[~teams["W"].str.contains("Division")]
```

```python
# Removing asterisks from team names

teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams.head()
```

```python
# Taking a look at team names in teams data versus players data

teams["Team"].unique()
```

```python
combined["Tm"].unique()
```

```python
# Creating dictionary for team nicknames

nicknames = {}

with open("../data/nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
nicknames
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head()
```

```python
# Merging team data with combined data

stats = combined.merge(teams, how="outer", on=["Team", "Year"])
```

```python
stats
```

```python
# Removing Unnamed 0 row

del stats["Unnamed: 0"]
```

```python
# Checking data types

stats.dtypes
```

```python
# Converting stats to numeric data

stats = stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats.dtypes
```

```python
# Checking why games back (GB) is not numeric

stats["GB"].unique()
```

```python
# Replacing '-' in games back with 0 as the dash signifies team is 0 games back

stats["GB"] = stats["GB"].str.replace("—", "0")
```

```python
stats["GB"].unique()
```

```python
# converting GB to numeric

stats["GB"] = pd.to_numeric(stats["GB"])
```

```python
stats.dtypes
```

```python
# Saving the combined data

stats.to_csv("../data/player_mvp_stats.csv")
```

```python
# Data exploration

# Highest scoring players

highest_scoring = stats[stats["G"] > 70].sort_values("PTS", ascending=False).head(10)
```

```python
highest_scoring.plot.bar("Player", "PTS")
```

```python
# Highest scorer each year

yearly_high_score = stats.groupby("Year").apply(lambda x: x.sort_values("PTS", ascending = False).head(1))
```

```python
yearly_high_score.plot.bar("Year", "PTS")
```
